### Corraltion heatmap
This notebook calculate and correlation of matrix and plot the heatmap plot

```python
import csv
import os
import pandas as pd
from venn import venn
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from lib.stats import stats
from lib.clean import clean
```

```python
df = pd.read_pickle("/global/cfs/cdirs/kbase/KE-Catboost/ziming/Enigma/GO/data/merged_std_nmlzed_common_GO.pkl")
df
```

```python
corr = df.T.corr()
corr
```

```python
merge_df_corr = stats.cluster_corr(corr,  distance_mathod="oneminus", threshold=0.2)
ax = sns.heatmap(round(merge_df_corr ,2),cmap='coolwarm', vmin=-1, vmax=1,# cmap='YlOrRd',\
                 xticklabels=merge_df_corr .index, \
                 yticklabels=merge_df_corr .index, \
                 annot=True, annot_kws={"size": 3})
fig = plt.gcf()
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
fig.set_size_inches(20.5, 18.5)
plt.show()
fig.savefig('Mgnify_and_Enigma_GO_clusterd_standard_normalized_correlation.png', dpi=500, bbox_inches='tight')
```

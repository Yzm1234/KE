```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
#from bokeh.io import show, output_file
#from bokeh.plotting import figure
```

```python
df = pd.read_pickle("/global/cfs/cdirs/kbase/KE-Catboost/ziming/Enigma/GO/data/merged_std_nmlzed_common_GO.pkl")
df.insert(0, 'label', df.index)
df.insert(1, 'dataset', ['Enigma' if 'EB271' in label else 'Mgnify' for label in df.index])
df
```

```python
# remove the outlier EB271-05-04
df = df[df['label'] != 'EB271-05-04']
df
```

```python
enigma_df = df[df['dataset'] == 'Enigma']
enigma_df
```

```python
mgnify_df = df[df['dataset'] == 'Mgnify']
mgnify_df
```

```python
# group enigma
avg_enigma = enigma_df.groupby(['dataset']).mean()
avg_enigma.insert(0, 'label', ['EB271'])
avg_enigma.insert(1, 'dataset', ['Enigma'])
avg_enigma
```

```python
# concatenate enigma and mgnify dataset
all_avg_df = pd.concat([avg_enigma, mgnify_df], axis=0)
all_avg_df
```

```python
# select biome of interest from mgnify to compare with enigma
target_label = ['root:Engineered:Food production:Fermented beverages',  'root:Engineered:Built environment',
                'root:Host-associated:Human:Skin', 'EB271',
                'root:Environmental:Aquatic:Freshwater:Pond', 'root:Host-associated:Porifera']
target_df = all_avg_df[all_avg_df['label'].isin(target_label)]
target_df = target_df.set_index(['label'])
del target_df['dataset']
target_df
```

```python
# find minimum value in dataset
cur_min = float('inf')
for row in target_df.to_numpy():
    for c in row:
        if c > 0 and c < cur_min:
            cur_min = c
cur_min
```

```python
target_df = target_df + 1e-8
target_df
```

```python
def prepare_data(data):
    ratio_df = pd.DataFrame(data, columns=['ratio'])
    # ratio_df_filter = ratio_df.loc[(ratio_df['ratio'] > 10) | (ratio_df['ratio'] < -10)]
    ratio_df_filter_sorted = ratio_df.sort_values(by=['ratio'])
    return ratio_df_filter_sorted
```

```python
subplot_idx = 0
fig, axs = plt.subplots(5, 3, figsize=(30,30))
for i in range(5):
    for j in range(i+1, 6):
        # subplot index
        subplot_x, subplot_y = subplot_idx//3, subplot_idx%3
        sample_1, sample_2 = target_df.index[i], target_df.index[j]
        data = np.log2(target_df.iloc[i]/target_df.iloc[j])
        processed_data = prepare_data(data)
        x_pos = [i for i, _ in enumerate(processed_data.index)]
        axs[subplot_x, subplot_y].bar(x_pos, processed_data.iloc[:,0], color='green')
        axs[subplot_x, subplot_y].set_title('{} vs {}'.format(sample_1, sample_2))
        subplot_idx += 1
fig.savefig('log_ratio_std.png')
```

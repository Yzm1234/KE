### The purpose of this file is to remove the duplicated rows in go_aggregated_runs_full.tsv where all column except run_id are the same.

```python
import csv
from tqdm.notebook import tqdm, trange
```

```python
with open('data/go_aggregated_runs_full.tsv', 'r') as in_file:
    with open('data/go_aggregated_runs_full_removed_duplicates.tsv', 'w') as out_file:
        reader = csv.reader(in_file, delimiter="\t")
        writer = csv.writer(out_file, delimiter="\t")
        header = next(reader)
        writer.writerow(header)
        last_row = next(reader)
        writer.writerow(last_row)
        for row in reader:
            # if current row has same cols as the last row, then remove current row
            if row[1:] == last_row[1:]:
                print("row with run_id({}), sample_id({}) is removed.".format(row[0], row[2]))
            else:
                writer.writerow(row)
                last_row = row
print("file generated!")
```

### Feature Selection
This notebook is applying correlation filter on raw data

```python
import pandas as pd
from sklearn import preprocessing
import os
import numpy as np
from lib.stats import stats
from lib.clean import clean
```

```python
data_dir = "/global/homes/z/zimingy/KE-Catboost/ziming/GO/"
output_path = "/global/cfs/cdirs/kbase/KE-Catboost/ziming/GO/data2/"
data_file = os.path.join(data_dir, 'data/go_aggregated_4.1/go_aggregated_4.1_mixed_updated.pkl')
```

#### 1. Read input file

```python
orig_df = pd.read_pickle(data_file)
orig_df = clean.rows_and_cols_quant_filter(orig_df, start_col_index=6, cutoff=0, pandas=True)
orig_df
```

```python
go_values = orig_df.iloc[:, 6:]
go_values
```

```python
# %%time
# pandas build in method takes too long
# corr_df = go_values.corr()
# corr_df
# 
go_values.to_numpy().shape
```

#### 2. Calculate correlation 2d array

```python
x = go_values.to_numpy().T
corr_np = stats.pearson_correlation_coefficient(x)
```

#### 3. Select features with low correlation (filter out highly correlated features)

```python
features_mask = stats.feature_extraction(corr_np, 0.9)
selected_features = go_values.columns[features_mask]
```

```python
processed_df = pd.concat([orig_df.iloc[:, :6], orig_df.loc[:, orig_df.columns.isin(selected_features)]], axis=1)
processed_df
```

#### 4. Save processed new files
In both tsv and pickle format

```python
output_path = "/global/cfs/cdirs/kbase/KE-Catboost/ziming/GO/data2/"
processed_df.to_csv(os.path.join(output_path, "go_aggregated_4.1_mixed_updated_feature_selected.tsv"), sep="\t", index=False)
processed_df.to_pickle(os.path.join(output_path, "go_aggregated_4.1_mixed_updated_feature_selected.pkl"))
```

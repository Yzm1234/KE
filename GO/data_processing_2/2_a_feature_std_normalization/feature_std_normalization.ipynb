## Feature standard normalization
This notebook using standard scaler to normalize data

```python
import os
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

#### 1. read input data file

```python
input_dir = "/global/cfs/cdirs/kbase/KE-Catboost/ziming/GO/data2"
input_df = pd.read_pickle(os.path.join(input_dir, "go_aggregated_4.1_mixed_updated_feature_selected.pkl"))
input_df
```

```python
values_array = input_df.iloc[:, 6:].to_numpy()
values_array.shape
```

#### 2. Data standard normalization

```python
scaler = preprocessing.StandardScaler().fit(values_array)
values_scaled  = scaler.transform(values_array)
input_df[input_df.columns[6:]] = values_scaled
input_df
```

#### 3. Save processed data to files

```python
input_df.to_csv("go_aggregated_4.1_mixed_updated_feature_selected_std_nmlz.tsv", sep="\t", index=False)
input_df.to_pickle("go_aggregated_4.1_mixed_updated_feature_selected_std_nmlz.pkl")
```
